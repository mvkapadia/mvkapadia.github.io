# M3: Advanced Matplotlib Graphing Assignment 

Importing matplotlib and pandas and executing the inline magic command to allow graphing in the Juypter notebook. 

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_ct= pd.read_csv('ConstructionTimeSeriesDataV2.csv')          #reading in the .csv file with pd and using the dataframe df_ct

fig,ax = plt.subplots()
ax.plot(df_ct['Month'],df_ct['Private Construction'])           #creating a line graph of Vehicle Sales over time

ax.xaxis.set_tick_params(which = 'both', top = False, bottom = True, labelbottom = True, direction='in')
ax.yaxis.set_tick_params(which = 'both', right = False, left = True, labelleft = True, direction='in')

#Enhance the axes object
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_xlabel('Months',fontsize='14',fontname='Times New Roman')
ax.set_ylabel('Private Construction (in dollars)',fontsize='14',fontname='Times New Roman')
ax.set_ylim(0,max(df_ct['Private Construction'])+10000)
ax.axvspan(74,110,ymin=0,ymax = 80000,facecolor='k',alpha=0.1)
ax.text(92,75000,'Change in Contruction\nTrend',fontsize=14,verticalalignment='center',horizontalalignment='center',fontname='Times New Roman')
ax.set_yticklabels(['{:,.0f}'.format(val) for val in ax.get_yticks()]) #There is a warning message that a FixLocator must be used which was present in the sample Jupyter notebook as well. 
#['{:,.0f}'.format(val) for val in ax.get_yticks()]                    #This line of code is to verify formatting the y ticks

#Enhance Figure Object
fig.set_size_inches(14,6)
fig.suptitle("Time Series Plot of Private Construction in the US",fontsize='20',fontname='Times New Roman')
fig.savefig('Private Construction in the US over time.jpg',dpi=600)
plt.show()
```
