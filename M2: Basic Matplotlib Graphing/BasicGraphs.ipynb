# Basic Matplotlib Graphing 

# Filename: Ozone.csv 

Creating some basic scatterplots from Ozone.csv data to test whether Temperature, Radiation or Wind would have the highest correlation or strongest relationship with Ozone Levels. 

```python
%matplotlib inline               
import pandas as pd              #importing pandas with the alias pd
import matplotlib.pyplot as plt  #importing matplot.lib.pyplot with the alias plt

df_oz =pd.read_csv('ozone.csv') #reading in the csv file via pd.read_csv

fig,ax = plt.subplots()
ax.scatter(df_oz['ozone'],df_oz['temperature']) #creating a scatterplot with Ozone Levels vs Temperature Levels. 
fig.suptitle ('Ozone Levels vs. Temperature Levels')
ax.xaxis.set_label_text('Ozone Levels')
ax.yaxis.set_label_text('Temperature Levels')
ax.set_ylim(0,150)  
fig.set_size_inches(7,5)
plt.show()
```

Analysis: When looking at the Ozone Levels vs the Temperature Levels we can see there is a strong relationship when the ozone levels are lower (0-50) and the temperature ranges from 50-100.
