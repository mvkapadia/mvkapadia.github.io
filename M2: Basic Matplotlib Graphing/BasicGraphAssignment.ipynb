# Do opioid overdoes tend to be associated with less affluent areas—that is, areas where families have lower incomes?

As scatterplots are useful for showing the relationship between 2 things as in this case with opioid overdoes and less affluent areas, it is the appropriate graph type.

```python
%matplotlib inline               
import pandas as pd              #importing pandas with the alias pd
import matplotlib.pyplot as plt  #importing matplot.lib.pyplot with the alias plt

df_opVA =pd.read_csv('OpidsVA.csv') #reading in the csv file via pd.read_csv

fig,ax = plt.subplots()
ax.scatter(df_opVA['MedianHouseholdIncome'],df_opVA['FPOO-Rate']) #creating a scatterplot with MedianHouseholdIncome vs Opiod Overdose Rate. 
fig.suptitle ('Opioid Overdoses vs. Median Household Income')
ax.xaxis.set_label_text('County Yearly Median Household Income')
ax.yaxis.set_label_text('Opioid Overdose Rate')
fig.set_size_inches(7,5)
plt.show()
```

# Some people who start with opioid addictions are reported to transition to heroin use. What is the relationship in Virginia counties between opioid overdoses and heroin overdoses?

As scatterplots are useful for showing the relationship between 2 things as in this case with Opioid overdoes and Heroine overdoses, it is the appropriate graph type.

```python
%matplotlib inline                      
import pandas as pd                     #importing pandas with the alias pd
import matplotlib.pyplot as plt         #importing matplot.lib.pyplot with the alias plt

df_opVA =pd.read_csv('OpidsVA.csv')     #reading in the csv file via pd.read_csv

fig,ax = plt.subplots()
ax.scatter(df_opVA['FPOO-Rate'],df_opVA['FFHO-Rate'])   #creating a scatterplot to plot the Opioid Overdose Rate vs the Heroine Overdose Rate
fig.suptitle ('Opioid Overdoses vs. Heroine Overdoses')
ax.xaxis.set_label_text('Heroine Overdose Rate')
ax.yaxis.set_label_text('Opioid Overdose Rate')
fig.set_size_inches(7,5)
plt.show()
```

# Filename: presidents.csv

# Which states are associated with the greatest number of United States presidents in terms of the presidents’ birthplaces?

The states associated with the greatest number of US presidents in terms of their birthplace is distributional data, and so a histogram is the most appropriate graph type. 

```python
%matplotlib inline                                    
import pandas as pd                                    #importing pandas with the alias pd
import matplotlib.pyplot as plt                        #importing matplot.lib.pyplot with the alias plt  

df_pres =pd.read_csv('presidents.csv')                 #reading in the csv file via pd.read_csv

state_dict = dict(df_pres['State'].value_counts())     #creating a dictionary,state_dict,for the frequency of Presidents' birth states

fig, ax = plt.subplots()
ax.bar(state_dict .keys(), state_dict .values())       #creating a frequency histogram data from DataFrame
fig.suptitle ("Frequency of US Presidents' Birth States")
ax.xaxis.set_label_text("US Presidents' Birth States")
ax.yaxis.set_label_text('Frequency of Birth States')
fig.set_size_inches(7,5)
plt.show()
```

# Filename: TOTALNSA.csv

# How have vehicle sales in the United States varied over time?

The vehicle sales in the United States over time is time series data, and so a line graph is the appropriate graph type.

```python
%matplotlib inline                                     
import pandas as pd                                    #importing pandas with the alias pd
import matplotlib.pyplot as plt                        #importing matplot.lib.pyplot with the alias plt  

df_tnsa=pd.read_csv('TOTALNSA.csv')                    #reading in the csv file via pd.read_csv

fig,ax = plt.subplots()
ax.plot(df_tnsa['DATE'],df_tnsa['TOTALNSA'])           #creating a line graph of Vehicle Sales over time
fig.suptitle('Vehicle sales in the US over time')
ax.xaxis.set_label_text('Date (in year-month-date)')
ax.yaxis.set_label_text('Total Sales')
ax.set_ylim(0,2000)                                    # Setting the min and max for y axis to have a Zero Basis graph 
fig.set_size_inches(7,5)
plt.show()
```
